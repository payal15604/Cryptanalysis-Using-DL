```python
import sys
from grain import ECRYPT_ctx, ECRYPT_keysetup, ECRYPT_ivsetup, ECRYPT_keystream_bytes

def printData(key, IV, ks):
    print("\n\nKey:        ", end="")
    for byte in key:
        print(f"{byte:02x}", end="")
    print("\nIV :        ", end="")
    for byte in IV:
        print(f"{byte:02x}", end="")
    print("\nkeystream:  ", end="")
    for byte in ks:
        print(f"{byte:02x}", end="")
    print()

def hex2data(hexstring):
    if len(hexstring) % 2 != 0:
        return None  # Hex string has an odd length

    data = bytearray()
    for i in range(0, len(hexstring), 2):
        hex_byte = hexstring[i:i+2]
        data.append(int(hex_byte, 16))

    return data

def main():
    key_hex = "0123456789abcdef1234"
    IV_hex = "0123456789abcdef"

    key = hex2data(key_hex)
    IV = hex2data(IV_hex)

    if key is None or IV is None:
        print("Invalid hex string.")
        sys.exit(1)

    ctx = ECRYPT_ctx()
    ks = bytearray(10)

    print("Grain")

    ECRYPT_keysetup(ctx, key, 80, 64)
    ECRYPT_ivsetup(ctx, IV)
    ECRYPT_keystream_bytes(ctx, ks, 10)
    printData(key, IV, ks)

if __name__ == "__main__":
    main()
```
